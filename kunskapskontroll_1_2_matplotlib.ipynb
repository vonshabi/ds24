### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 100)
y = x**2

plt.plot(x, y)
plt.title("y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show() 

#OOapproach
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("y = x^2")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(True)
plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
#create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

#scatter plot
ax1.scatter(random_data_x, random_data_y, alpha=0.3, 
label='Random Data')
ax1.plot(x, y, color='red', label='y = y^2')
ax1.set_title('Scatter Plot')
ax1.set_xlabel('X values')
ax1.set_ylabel('Y values')
ax1.legend()

#bar plot
ax2.bar(names, values, color='pink')
ax2.set_title('Bar Plot Of Fruit Counts')
ax2.set_ylabel('Count')
ax2.set_xlabel('Fruit Types')

plt.tight_layout()
plt.show()
```
